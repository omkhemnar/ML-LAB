```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('used_cars_data.csv')

x = data['Name'].head(10)
y = data['Price'].head(10)

plt.bar(x, y, color='teal')
plt.xlabel('Car Name')
plt.ylabel('Price (in Lakhs)')
plt.title('Top 10 Used Car Prices')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('used_cars_data.csv')

fuel_counts = data['Fuel_Type'].value_counts()

plt.figure(figsize=(6, 6))  # Increase figure size
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 10})
plt.title('Distribution of Cars by Fuel Type')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('used_cars_data.csv')

data['Year'] = pd.to_numeric(data['Year'], errors='coerce')
year_data = data['Year'].dropna()

plt.hist(year_data, bins=15, color='orange', edgecolor='black')
plt.xlabel('Manufacturing Year')
plt.ylabel('Number of Cars')
plt.title('Distribution of Cars by Year')
plt.tight_layout()
plt.show()
```
